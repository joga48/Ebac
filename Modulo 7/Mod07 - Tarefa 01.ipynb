````python
dftp = df.dtypes
################
dftp
````

````python
lst = []
dflst = []
for i in df:
    lst = []
    lst.append(df[i][0])
    for j in df[i]:
        ad = True
        for k in lst:
            if j == k:
                ad = False
        if ad:
            lst.append(j)
    dflst.append(len(lst))
dftp = pd.DataFrame(dftp)
dftp['qtd_categorias'] = dflst
dftp = dftp.rename(columns = {0:'tipo'} )
dftp
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
dfnew = df[['qtd_filhos','idade','tempo_emprego',
            'possui_celular','possui_fone_comercial',
            'possui_fone','possui_email','qt_pessoas_residencia','mau']]
dfnew = dfnew.merge(pd.get_dummies(df['sexo']), left_index = True, right_index = True)
dfnew = dfnew.merge(pd.get_dummies(df['educacao']), left_index = True, right_index = True)
dfnew = dfnew.merge(pd.get_dummies(df['estado_civil']), left_index = True, right_index = True)
dfnew = dfnew.merge(pd.get_dummies(df['tipo_residencia']), left_index = True, right_index = True)
dfnew = dfnew.merge(pd.get_dummies(df['tipo_renda']), left_index = True, right_index = True)
dfnew = dfnew.merge(pd.get_dummies(df['posse_de_veiculo']), left_index = True, right_index = True)
dfnew = dfnew.merge(pd.get_dummies(df['posse_de_imovel']), left_index = True, right_index = True)
dfnew
````
