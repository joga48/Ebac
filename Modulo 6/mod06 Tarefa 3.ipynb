#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
df.loc[df['APGAR5'] <= 3, 'apgar5_cat'] = 'asfixia_severa'
df.loc[df['APGAR5'] >= 4, 'apgar5_cat'] = 'asfixia_moderada'
df.loc[df['APGAR5'] >= 6, 'apgar5_cat'] = 'asfixia_leve'
df.loc[df['APGAR5'] >= 8, 'apgar5_cat'] = 'normal'
df
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
crz = pd.crosstab(df['GESTACAO'], df['APGAR5'])
crz
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
#save = pd.DataFrame(index=['22 a 27 semanas', '28 a 31 semanas', '32 a 36 semanas', '37 a 41 semanas', '42 semanas e mais', 'Menos de 22 semanas'])
sum = crz.sum()
crz_dup = crz
crz_dup = crz_dup / sum
crz_dup
```
